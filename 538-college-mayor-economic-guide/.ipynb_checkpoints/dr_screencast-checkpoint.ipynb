```python
pdf = df.nlargest(
    n=20,
    columns="median",
)[["major", "major_category", "median", "p25th", "p75th"]].melt(
    id_vars=["major", "major_category"],
    var_name="earning_percentile",
    value_name="earning",
)

# hack, when melting i have the quantiles:
# errorbar=lambda x: (x.min(), x.max()) = ('pi', 100)
# estimator=median
sns.pointplot(
    # mappings
    data=pdf,
    x="earning",
    y="major",
    hue="major_category",
    # hack
    estimator="median",
    errorbar=lambda x: (x.min(), x.max()),
    # aesthetic, some sns, some plt
    markers="D",
    markersize=5,
    linestyles="none",  # will conect the markers, too crowded
    capsize=0.4,
    linewidth=1.2,
    fillstyle="none",
).set(
    title="Inter quartile range of Earnings by Top 20 Major",
    ylabel="",
)
```

### Exploring sample sizes

Considering an approach based on [Bland-Altman Plot](https://en.wikipedia.org/wiki/Bland%E2%80%93Altman_plot).

- Scatter plot, y: variable of interest, x: totals


```python
df.plot.scatter(x="sample_size", y="median").set(xscale="log")
```

- There is a lot more variation in the median ernings in the low sample sizes (<50)
- Picking a cut of: 100


```python
cut_off_sample_size = 100
pdf = (
    df[df["sample_size"] > cut_off_sample_size]
    .nlargest(
        n=20,
        columns="median",
    )[["major", "major_category", "median", "p25th", "p75th"]]
    .melt(
        id_vars=["major", "major_category"],
        var_name="earning_percentile",
        value_name="earning",
    )
)
plt.figure(figsize=(12, 8))
# hack, when melting i have the quantiles:
# errorbar=lambda x: (x.min(), x.max()) = ('pi', 100)
# estimator=median
sns.pointplot(
    # mappings
    data=pdf,
    x="earning",
    y="major",
    hue="major_category",
    # hack
    estimator="median",
    errorbar=lambda x: (x.min(), x.max()),
    # aesthetic, some sns, some plt
    markers="D",
    markersize=5,
    linestyles="none",  # will conect the markers, too crowded
    capsize=0.4,
    linewidth=1.2,
    fillstyle="none",
).set(
    title=f"Inter quartile range of Earnings by Top 20 Majors with at least {cut_off_sample_size} graduates surveyed",
    ylabel="what are the highest earning majors",
    xlabel="Graduate Earnings",
)
```

- One should be really carefull. If one will be doing a s statistical test one should not be doing that kind of slicing, in order to very carefull doing in advance because you can be p-hacking.
- But if it is more related to communicate trhough a visualization, is more about showing what your audience is interested in. They might not be interested in the most obscure majors (ones with low sample size)

```python
(
    df[['major_category', 'total', 'men', 'women']]
    .groupby('major_category')
    .sum()
    .sort_values('total')
    .plot.barh(y=['men', 'women'], stacked=True)
    .set(title='Top 20 major categories by gender')
)
```

```python
# Pandas warnings explained
# https://galea.medium.com/pandas-groupby-with-multiple-columns-c7a94ef0be86
# include_groups=False: https://stackoverflow.com/questions/77969964/deprecation-warning-with-groupby-apply/78030503#78030503

pdf = (
    df
    .groupby('major_category').apply(lambda x: pd.Series({
        'total': x['total'].sum(),
        'men': x['men'].sum(),
        'women': x['women'].sum(),
        'share_women': x['women'].sum() / x['total'].sum(),
        'weighted_median_salary': (
            (x['median'] * x['sample_size']).sum()/(x['sample_size'].sum())
        ).round(2)
    }), include_groups=False)
    .reset_index()
)

# pdf.plot(
#     x='share_women', y='weighted_median_salary',c='major_category', colormap='viridis'
# )

sns.set_theme()
sns.regplot(data=pdf,
    x='share_women', y='weighted_median_salary',
    # hue='major_category'
).set(ylim=0)

# add annotations one by one with a loop
for line in range(0, pdf.shape[0]):
     plt.text(
        pdf['share_women'][line], 
        pdf['weighted_median_salary'][line], 
        pdf['major_category'][line], 
        horizontalalignment='left', 
        size='6', color='black'
    )
```

To consider prior statistics:

- If you are applying a statistical test **you probably don't whant to aggregate first**, because if you do you are going to lose a lot of the information about variation that you have. 
- It will also keep us safer from [Simpson's Paradox](https://www.youtube.com/watch?v=sxYrzzy3cq8): the same set of data can appear to show opposite trends depending on how it's grouped. This often occures when **aggregated data hides a conditional variable (a lurking variable)**, which is a hidden additional factor that significantly influences results. There's no one unique way to avoid this paradox, data can be divided and group in different ways, but overall numbers (all factors considered) may sometimes give a more accurate picture than data divided into misleading or arbitrary categories. **All we can do is carefully study the actual situations the statistics describe and considering whether lurking variables might  be present**. [another case by minute physics](https://www.youtube.com/watch?v=E_ME4P9fQbo), [a practical example of a/b testing](https://www.youtube.com/watch?v=F3mEQ1jh04M&t=188s), [covid scenario and good examples](https://www.youtube.com/watch?v=t-Ci3FosqZs).

Sympson's Paradox in the case of Health category (share women vs earnings)

- We saw that **Health** category has the highest share women (more women graduate in that major) and they also have relatively high median salary
- We also saw that **Nursing** was among the highest in share women but not that high in salaries compared to other majors. 
- So, it might be that areas within Health that have more women also have lower salaries. 
- It could be that correlations (share women vs earning) within each category that are different than the overall correlation.

- In fact, haciendo el plot anterior solo a Health la correlación cambia ligeramente, se hace ligeramente positiva a comparacion de cuando se agrego la data.
- It doesn't mean that one is better over the other (summaries at aggregated level vs overall), but one should be carefull to generalize conclusions regarding gender.
- Además, share_woman (the agregated one by category) is already a summary of a summary. Más además, it is a ratio of women/total, in which total is different across majors (en consecuencia across major categories).

So:

- Take care when deriving conclusions out of statistics of summaries. Statistics in general.

Lurking vs Confounding:

- Lurking variables are variables that have not been considered in the study, but are influencing both studied variables. 

- Confounding variables are variables that were considered in the study and influence one variable, but the confounding variable wasn't credited with the relationship.
